# Question Explainer

This notebook uses the Gemini API (via the OpenAI-compatible interface) to answer questions in a detailed, step-by-step way. The response is streamed and rendered as Markdown directly in the notebook.

## Imports

We start by importing the libraries we need. This includes `dotenv` for loading environment variables, `IPython.display` for rendering Markdown output, and the `OpenAI` client which we will point at the Gemini API.

```python
import os
import json
from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display
from openai import OpenAI
```

## Setting Up the API Client

Here we load the API key from a `.env` file, set the base URL to point at the Gemini API, choose the model, and create the client.

```python
load_dotenv(override=True)
api_key = os.getenv('GOOGLE_API_KEY')
GEMINI_BASE_URL = "https://generativelanguage.googleapis.com/v1beta/openai/"
MODEL = 'gemini-2.5-flash'

gemini = OpenAI(api_key=api_key, base_url=GEMINI_BASE_URL)
```

## Defining the System Prompt

The system prompt tells the model how to behave. We instruct it to simplify questions, break them into smaller parts, and walk through the solution with bullet points.

```python
system_propmt = """
You are an assistant that helps with answering questions and explaining them
in full details.
Simplify the question and break it into pieces and come up with a helpful solution.
Walk through the solution with bullet points.
"""
```

## Creating the User Prompt

This function takes a question string and wraps it into a simple user prompt that gets sent to the model.

```python
def user_prompt(question: str):
    prompt = "Answer this questions please: \n"
    return prompt + question
```

A quick test to make sure the prompt looks right before sending it to the model.

```python
print(user_prompt("What is an LLM"))
```

## Building the Message List

The `messages` function puts together the list of messages in the format the API expects, with a system message and a user message.

```python
def messages(question):
    return [
        {"role":"system", "content":system_propmt},
        {"role":"user", "content":user_prompt(question)}
    ]
```

## Streaming the Response

The `answer` function sends the question to the Gemini API with streaming enabled. As chunks of the response come in, the display is updated in real time so you can see the answer being written out progressively.

```python
def answer(question):
    stream = gemini.chat.completions.create(
        model=MODEL,
        messages=messages(question),
        stream=True
    )
    response = ""
    display_handle = display(Markdown(""), display_id=True)
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        update_display(Markdown(response), display_id=display_handle.display_id)
```

## Asking a Question

Now we can call the `answer` function with any question and see the response streamed below.

```python
answer("What is an LLM")
```
